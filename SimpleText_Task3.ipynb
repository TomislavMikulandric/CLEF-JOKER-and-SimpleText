```python
import re

def remove_redundant_text(text):
    # Define patterns to search for
    patterns = [
        r'^Hey there!',
        r'^Sure!',
        r'^As a scientific journalist,',
        r'I\'m here to break down a complex study into simple terms for you\.',
        r'Here\'s a simplified version of the text',
        r'Let me break it down for you:',
        r'I\'m here to break down a complex study into simple terms for you\.',
        r'I\'m here to break down complex scientific concepts into simple, easy-to-understand language.',
        r'I\'m here to break down a complex topic into simpler terms for you. So, let\'s talk about',
        r'Sure, I\'d be happy to help!',
        r'Here\'s a simplified explanation of',
        r'In other words,',
        r'In simple terms,'

    ]
    # Compile regular expressions
    regex_patterns = [re.compile(pattern) for pattern in patterns]

    # Remove patterns from text
    for pattern in regex_patterns:
        text = re.sub(pattern, '', text).strip()

    return text
```

```python
snt_df['simplified_snt'] = snt_df['simplified_snt'].apply(remove_redundant_text)
snt_df
```

```python
abs_df['simplified_abs'] = abs_df['simplified_abs'].apply(remove_redundant_text)
abs_df
```

```python
def format_abs_row_to_json(row):
    json_obj = {
        "run_id": "Tomislav&Rowan_Task3.1_LLAMA",  # Update with your team_id, task_id, and method_used
        "manual": 0,  # Update based on whether the run is manual or not
        "abs_id": row["abs_id"],  # Assuming 'abs_id' is used for Task 3.2
        "simplified_abs": row["simplified_abs"]
    }
    return json_obj

def format_snt_row_to_json(row):
    json_obj = {
        "run_id": "Tomislav&Rowan_Task3.1_LLAMA",  # Update with your team_id, task_id, and method_used
        "manual": 0,  # Update based on whether the run is manual or not
        "snt_id": row["snt_id"],  # Assuming 'snt_id' is used for Task 3.1
        "simplified_snt": row["simplified_snt"]
    }
    return json_obj
```

```python
import json

# Iterate over each row and format it into a JSON object
output_abs_json = []
output_snt_json = []
for index, row in abs_df.iterrows():
    output_abs_json.append(format_abs_row_to_json(row))

# Write the JSON output to a file
with open('abs_output.json', 'w') as f:
    json.dump(output_abs_json, f, indent=4)

for index, row in snt_df.iterrows():
    output_snt_json.append(format_snt_row_to_json(row))

# Write the JSON output to a file
with open('snt_output.json', 'w') as f:
    json.dump(output_snt_json, f, indent=4)
```
